## Autoencoder

```python
from keras.layers import Input, Dense
from keras.models import Model
from keras.optimizers import Adam

def create_autoencoder(hidden_units):
    input_layer = Input(shape=(X_train.shape[1],))
    encoded = Dense(hidden_units, activation='relu')(input_layer)
    decoded = Dense(X_train.shape[1], activation='sigmoid')(encoded)
    
    autoencoder = Model(input_layer, decoded)
    
    autoencoder.compile(optimizer=Adam(), loss='mae')
    
    return autoencoder

hidden_units = [8, 16, 32]
best_autoencoder = None
best_loss = float('inf')

for units in hidden_units:
    autoencoder = create_autoencoder(units)
    
    history = autoencoder.fit(
        X_train, X_train,
        epochs=100,
        batch_size=24,
        shuffle=True,
        validation_data=(X_test, X_test),
        verbose=1 
    )
    
    val_loss = np.min(history.history['val_loss'])
    if val_loss < best_loss:
        best_loss = val_loss
        best_autoencoder = autoencoder

print("Best validation loss: ", best_loss)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Define function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, classes, cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    """
    conf_matrix = confusion_matrix(y_true, y_pred)
    
    fig, ax = plt.subplots(figsize=(8, 6))
    im = ax.imshow(conf_matrix, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    
    ax.set(xticks=np.arange(conf_matrix.shape[1]),
           yticks=np.arange(conf_matrix.shape[0]),
           xticklabels=classes, yticklabels=classes,
           title="Confusion Matrix",
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = 'd'
    thresh = conf_matrix.max() / 2.
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            ax.text(j, i, format(conf_matrix[i, j], fmt),
                    ha="center", va="center",
                    color="white" if conf_matrix[i, j] > thresh else "black")
    fig.tight_layout()
    plt.show()

# Create true and predicted labels
true_labels = [0] * len(X_test)  # since all are not anomalies
predicted_labels = [1 if loss > threshold else 0 for loss in test_mae_loss]

# Call the function
plot_confusion_matrix(true_labels, predicted_labels, classes=["Not Anomaly", "Anomaly"])
```

```python
import os
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt

def detect_anomalies(best_autoencoder, dataset, save_path, threshold=threshold):
    
    # Select only the columns listed in feature_list
    features = dataset[feature_list]
    
    # Ensure that the features are of numeric type
    features.dropna(inplace=True)

    features = scaler.transform(features)

    X_pred = best_autoencoder.predict(features)
    
    # Calculate MAE loss
    mae_loss = np.mean(np.abs(X_pred - features), axis=1)

    # Plotting MAE loss
    plt.hist(mae_loss, bins=50)
    plt.axvline(x=threshold, color='red') # Draw a red vertical line at the threshold
    plt.xlabel("MAE loss")
    plt.ylabel("No of samples")
    
    # Save the plot to the specified directory with the feature number
    plt.savefig(os.path.join(save_path, "loss_distribution.png"))

    plt.show()
    plt.close()  # Close the current figure to free memory

    # Detect anomalies
    anomalies = (mae_loss > threshold).tolist()
    anomalies_indices = [index for index, is_anomaly in enumerate(anomalies) if is_anomaly]

    # Return DataFrame containing anomaly points
    anomaly_points = dataset.iloc[anomalies_indices]
    return anomaly_points
```

```python
import matplotlib.pyplot as plt
import os

def visualize_anomalies(anomaly_points, dataset, feature_list, save_path):
    # Create the directory if it does not exist
    os.makedirs(save_path, exist_ok=True)
    
    # Get the unique dates from anomaly_points
    unique_dates = anomaly_points['date'].unique()
    
    # Iterate through each feature to visualize
    for feature_idx, feature in enumerate(feature_list):
        if feature != "hour":
            plt.figure(figsize=(20, 10))

            # Iterate through each date
            for date in unique_dates:
                # Extract data for the specific date
                date_data = dataset[dataset['date'] == date]
                plt.plot(date_data['hour'], date_data[feature], label=f"Date: {date}")
                plt.scatter(date_data['hour'], date_data[feature])  # Scatter points for each hour

                # Highlight the anomaly points for this feature on this date
                anomaly_points_date = anomaly_points[anomaly_points['date'] == date]
                for index, row in anomaly_points_date.iterrows():
                    hour = row['hour']
                    value = row[feature]
                    plt.scatter(hour, value, color='red', s=100) # Point in red, size=100

            plt.xlabel('Hour')
            plt.ylabel(feature)
            plt.title(f"Feature: {feature}")
            plt.xlim(0, 23)
            plt.legend()

            # Save the plot to the specified directory with the feature number
            plt.savefig(os.path.join(save_path, f"{feature_idx + 1}.png"))
            plt.close()  # Close the current figure to free memory
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate the normal_ranges outside of the function
normal_ranges = df_normal_date.groupby(["nominal_idx", "hour"]).agg(['min', 'max'])

def visualize_anomalies_for_date(anomaly_points, dataset, feature_list, date_selected, normal_ranges):
    # Filter out "hour" from the feature list for plotting purposes
    features_to_plot = [feature for feature in feature_list if feature != "hour"]
    
    # Create a single long figure with subplots for features_to_plot
    fig, axes = plt.subplots(len(features_to_plot), 1, figsize=(9, 3 * len(features_to_plot)))

    # Ensure that axes is always a list for consistency in indexing
    if len(features_to_plot) == 1:
        axes = [axes]

    # Extract data for the specific date
    date_data = dataset[dataset['date'] == date_selected]
    anomaly_points_date = anomaly_points[anomaly_points['date'] == date_selected]

    # Determine the nominal_idx for the selected date
    nominal_idx_ranges = {
        1: ('2020-08-19', '2020-09-18'),
        2: ('2020-09-19', '2020-11-14'),
        3: ('2020-11-15', '2020-12-11')
    }
    date_selected_dt = pd.to_datetime(date_selected)
    for idx, (start, end) in nominal_idx_ranges.items():
        if pd.to_datetime(start) <= date_selected_dt <= pd.to_datetime(end):
            nominal_idx_for_date = idx
            break

    # Iterate through each feature to visualize
    for feature_idx, (ax, feature) in enumerate(zip(axes, features_to_plot)):
        ax.plot(date_data['hour'], date_data[feature], label=f"Date: {date_selected}")
        ax.scatter(date_data['hour'], date_data[feature])  # Scatter points for each hour

        # Highlight the anomaly points for this feature on this date
        for index, row in anomaly_points_date.iterrows():
            hour = row['hour']
            value = row[feature]
            ax.scatter(hour, value, color='red', s=100)  # Point in red, size=100

        # Shade the region for normal range values hour by hour
        for hour in range(24):
            if (nominal_idx_for_date, hour) in normal_ranges.index:  # Add this check
                feature_min = normal_ranges.loc[(nominal_idx_for_date, hour), (feature, 'min')]
                feature_max = normal_ranges.loc[(nominal_idx_for_date, hour), (feature, 'max')]
                ax.fill_between([hour, hour+1], feature_min, feature_max, color='green', alpha=0.1)

        ax.set_ylabel(feature)
        ax.set_xlim(0, 23)
        ax.legend()
        if feature_idx == len(features_to_plot) - 1:  # If it's the last subplot
            ax.set_xlabel('Hour')

    plt.tight_layout()
    plt.show()
```
